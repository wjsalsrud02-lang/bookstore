```python
# csv 파일 저장된 경로
csv_path = 'sample_sales.csv'
```

```python
# 파일 읽기
df = pd.read_csv(csv_path)
df
```

```python
# csv 파일을 저장'할' 경로
csv_path = 'sample_sales(1).csv'
```

```python
# 파일 저장
df.to_csv(csv_path)
```

```python
# 삼각김밥 데이터만 필터링
df_gimbap = df[df['product_name'] == '삼각김밥'].copy()
```

```python
# stock 컬럼 추가 (처음 재고 100개)
df_gimbap['stock'] = 100
```

```python
# stock-- 컬럼: stock에서 판매 수량(quantity)을 뺀 값
df_gimbap['stock--'] = df_gimbap['stock'] - df_gimbap['quantity']
```

```python
# 결과 미리보기
print("✅ 삼각김밥 데이터 미리보기:")
print(df_gimbap[['product_name', 'quantity', 'stock', 'stock--']].head())
```

```python
# stock 컬럼 기본값 0으로 초기화
df['stock'] = 0
df['stock--'] = 0
```

```python
# 삼각김밥 행만 처리
mask = df['product_name'] == '삼각김밥'
```

```python
# 삼각김밥은 재고 100개로 시작
df.loc[mask, 'stock'] = 100
```

```python
# 판매 수량만큼 차감한 재고
df.loc[mask, 'stock--'] = df.loc[mask, 'stock'] - df.loc[mask, 'quantity']
```

```python
# 결과 미리보기
print("✅ 삼각김밥 데이터 미리보기:")
print(df.loc[mask, ['product_name', 'quantity', 'stock', 'stock--']].head())
```
